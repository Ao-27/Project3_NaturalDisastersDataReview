```python
Filtered_Complete_DataAnalysis_df['disaster'] = 0

for rows in Filtered_Complete_DataAnalysis_df:
    if Filtered_Complete_DataAnalysis_df['Disaster Type'].any():
        Filtered_Complete_DataAnalysis_df['disaster'] = 1
        
Filtered_Complete_DataAnalysis_df
```

```python
disaster_df = Filtered_Complete_DataAnalysis_df.pivot_table(values='disaster', 
                                            index = Filtered_Complete_DataAnalysis_df.index, columns = 'Disaster Type')
```

```python

disaster_df
```

```python
labels2 = disaster_df.columns.tolist()
labels2
```

```python
labels = ['Dis_no', 'Country', 'Continent', 'Start_Year', 'Total_deaths', 'Total_Damages']


df = pd.DataFrame({labels[0]: Filtered_Complete_DataAnalysis_df['Dis No'],
                   labels[1]: Filtered_Complete_DataAnalysis_df['Country'],
                   labels[2]: Filtered_Complete_DataAnalysis_df['Continent'],
                   labels[3]: Filtered_Complete_DataAnalysis_df['Start Year'],
                   labels[4]: Filtered_Complete_DataAnalysis_df['Total Deaths'],
                   labels[5]: Filtered_Complete_DataAnalysis_df['Total Damages (\'000 US$)']
})



df
```

```python
df2  = df.join(disaster_df)
```

```python
df2
```

```python


countries = df2['Country'].unique()

#countries = np.delete(countries, np.where(countries == 'Argentina'))

countries
```

```python
sum_col = df2['Flood'].sum()
sum_col
```

```python
done = []
for country in countries:
    interim = df2.loc[df2['Country'] == country]
    
    deaths = interim['Total_deaths'].sum()
    damages = interim['Total_Damages'].sum()
    animal_acc = interim['Animal accident'].sum()
    drought = interim['Drought'].sum()
    earthquake = interim['Earthquake'].sum()
    epidemic = interim['Epidemic'].sum()
    ex_temp = interim['Extreme temperature '].sum()
    flood = interim['Flood'].sum()
    glacial = interim['Glacial lake outburst'].sum()
    impact = interim['Impact'].sum()
    insect = interim['Insect infestation'].sum()
    lanslide = interim['Landslide'].sum()
    mass_move = interim['Mass movement (dry)'].sum()
    storm = interim['Storm'].sum()
    volcano = interim['Volcanic activity'].sum()
    wildfire = interim['Wildfire'].sum()
    total_dis = animal_acc + drought + earthquake + epidemic + ex_temp + flood +glacial+impact+insect+lanslide+mass_move+storm+volacno+wildfire
    
    kinda_final = {
        'country': country,
        'Total_deaths': deaths,
        'Total_Damages': damages,
        'animal_acc': animal_acc,
        'drought': drought,
        'earthquake': earthquake,
        'epidemic': epidemic,
        'ex_temp': ex_temp,
        'flood': flood,
        'glacial': glacial,
        'impact': impact,
        'insect': insect,
        'lanslide': lanslide,
        'mass_move': mass_move,
        'storm': storm,
        'volcano': volcano,
        'wildfire': wildfire,
        'total_dis': total_dis
        }
    
    done.append(kinda_final)

done
 
```

```python
final = pd.DataFrame(done)
final
```

```python
#cleaning

final.loc[final['country'] == 'Comoros (the)', 'country'] = 'Comoros'
final.loc[final['country'] == 'Iran (Islamic Republic of)', 'country'] = 'Iran'
final.loc[final['country'] == 'Niger (the)', 'country'] = 'Niger'
final.loc[final['country'] == 'Korea (the Republic of)', 'country'] = 'South Korea'
final.loc[final['country'] == 'Philippines (the)', 'country'] = 'Philippines'
final.loc[final['country'] == 'United States of America (the)', 'country'] = 'United States'
final.loc[final['country'] == 'United Kingdom of Great Britain and Northern Ireland (the)', 'country'] = 'United Kingdom'
final.loc[final['country'] == 'Micronesia (Federated States of)', 'country'] = 'Micronesia'
final.loc[final['country'] == 'Venezuela (Bolivarian Republic of)', 'country'] = 'Venezuela'
final.loc[final['country'] == 'Dominican Republic (the)', 'country'] = 'Dominican Republic'
final.loc[final['country'] == 'Russian Federation (the)', 'country'] = 'Russian Federation'
final.loc[final['country'] == 'Korea (the Democratic People\'s Republic of)', 'country'] = 'North Korea'
final.loc[final['country'] == 'Moldova (the Republic of)', 'country'] = 'Moldova'
```
