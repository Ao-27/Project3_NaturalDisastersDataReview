```python
import requests
import json
import pandas as pd

#via URL => https://stackoverflow.com/questions/61633719/can-you-pull-an-api-call-straight-into-a-pandas-dataframe#:~:text=you%20can%20use%20json_normalize%3A%20import%20requests%20import%20pandas,%28%29%20df%20%3D%20json_normalize%20%28available_data%29%20print%20%28df.head%20%28%29%29

url = r'https://restcountries.com/v3.1/all'
data = requests.get(url)

#
# Store the API response in a variable.
available_data = data.json()

country_api_df1 =  pd.read_json(json.dumps(available_data))   #pd.json_normalize(available_data)

country_api_df1.head()
```
