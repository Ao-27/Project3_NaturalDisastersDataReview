```python
#if start month/day is NaN ... then "null" the date ... 
#via URL => https://stackoverflow.com/questions/17071871/how-do-i-select-rows-from-a-dataframe-based-on-column-values

#USE CASE => Became very valuable to create the "missing Y/N" column in the final DF with "Date_Completed" ... noted as a column based on the "joinability" ( IE: result of pd.Merge() )

Starting_missing_dates = Disaster_starting_dates_df.loc[(Disaster_starting_dates_df['Start Month'].isnull()) | Disaster_starting_dates_df['Start Day'].isnull()]

Starting_Rename_Missing = {
     'Start Year' : 'Missing_start_DF_Year',
     'Start Month' : 'Missing_start_DF_Month',
     'Start Day' : 'Missing_start_DF_Day'
}

Starting_missing_dates = Starting_missing_dates.rename(columns=Starting_Rename_Missing)

Starting_missing_dates
```
